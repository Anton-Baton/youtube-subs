```python
import youtube_dl
import ffmpy
import subprocess
import boto3
import os
```

```python
options = {
    'format': 'bestvideo[maxheight <=? 480][ext=webm]/best[ext=webm][max_height <=? 480]',
}
```

```python
def get_video_info(video_id, fields=['url', 'duration']):
    # TODO: handle errors
    url = 'https://www.youtube.com/watch?v={video_id}'.format(video_id=video_id)
    with youtube_dl.YoutubeDL(options) as ydl:
        res = ydl.extract_info(url, download=False)
    return {k: res[k] for k in fields}
```

```python
def get_video_preview(url, start, preview_duration=5):
    # TODO: handle errors
    ff = ffmpy.FFmpeg(
        inputs={url: '-ss {}'.format(start)},
        outputs={'pipe:1': ('-t 5 ' # duration of cutted snippet 
                            '-codec:v libvpx ' # video codec
                            '-quality realtime -cpu-used 0 ' # processing options
    #                                         '--threads 4 ' # Number of threads - set to the number of CPU cores
                            #'-b:v 500k ' # desired output video bitrate
                            #'-qmin 10 -qmax 42 ' # quantization levels (0-63) - lower better
                            #'-maxrate 500k ' 
                            #'-bufsize 1M ' # twice as max rate to have buffer for 2 seconds
                            '-vf scale=-1:480 ' # new video size (-1 means save proportions)
                            #'-codec:a libvorbis -b:a 128k ' # audio codec
                            '-f webm ')} # used to specify encoding from pipe
    )
    f, err = ff.run(stdout=subprocess.PIPE)
    return f
```

```python
def fetch_preview(video_id):
    info = get_video_info(video_id)
    url, duration = info['url'], info['duration']
    preview_start = 10 if duration > 20 else duration / 4
    preview_duration = 5 if duration > 10 else duration / 2
    return get_video_preview(url, preview_start, preview_duration)
    
```

```python
S3_BUCKET_NAME = 'youtubesubs'
S3_PREVIEWS_FOLDER = 'previews'
S3_BASE_URL = 'https://s3.eu-central-1.amazonaws.com'
def save_preview(filename, preview):
    # TODO: require secrets in env
    s3 = boto3.resource('s3')
    s3.Object(S3_BUCKET_NAME, os.path.join(S3_PREVIEWS_FOLDER, filename)).put(Body=preview)
    return os.path.join(S3_BASE_URL, S3_BUCKET_NAME, S3_PREVIEWS_FOLDER, filename)
```

```python
import time
def generate_filename():
    # TODO: use video_id ?
    return str(int(time.time()))+'.webm'
```

```python
def create_preview(video_id):
    preview = fetch_preview(video_id)
    filename = generate_filename()
    return save_preview(filename, preview)
```
